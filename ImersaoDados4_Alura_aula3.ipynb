````python
# 1 - Criar uma escala de valores em milhoes (10^6)

dados_vendas['Valor_anuncio_milhoes'] = dados_vendas['Valor_anuncio_float'].apply(lambda i: i/(1000000))
dados_vendas.head()
````

````python
# 2 - Deixar o gráfico do histograma de valores legível (alterar labels, cores, título, escala)

plt.figure(figsize=(15, 8))
ax = srn.histplot(data = dados_vendas, x = "Valor_anuncio_milhoes", kde = True)
ax.set_title("Histograma Valor Imóvel (milhões)")
plt.xlim((-5, 20))
plt.xlabel('Valor (milhões)')
plt.ylabel('count')
plt.show()
````

````python
# 3 - Preço do metro quadrado por bairro e plotar em um gráfico ideal

dados_vendas["Valor_m2"] = (dados_vendas['Valor_anuncio_float'])/dados['Metragem']
dados_vendas.head()
````

````python
plt.figure(figsize=(10, 6))
ax = srn.histplot(data = dados_vendas, x = "Valor_m2", kde = True)
ax.set_title("Histograma Preço por m²")
plt.xlim((-5, 40000))
plt.xlabel('Preço (m²)')
plt.ylabel('count')
plt.show()
````

# Imersão Dados - Alura - Aula 03

````python
dados_vendas.head()
````

````python
# calculando valor do m² por Bairro 
dados_vendas['Valor_anuncio_float'] / dados_vendas['Metragem']
````

````python
# adicionando uma coluna com o resultado do cálculo valor do m² por Bairro

dados_vendas["Valor_m2"] = dados_vendas["Valor_anuncio_float"]/dados_vendas["Metragem"]
dados_vendas.head()
````

````python
# media do valor do m² por bairro

dados_vendas.groupby('Bairro')['Valor_m2'].mean()
````

````python
# Agrupando por bairro

dados_vendas.groupby("Bairro").mean()
````

````python
# gerando um novo DataFrame(dados_bairro) com o retorno da soma dos valores das colunas do DataFrame(dados_vendas)

dados_bairro = dados_vendas.groupby("Bairro").sum()
dados_bairro
````

````python
# após a soma de todos os valores, foi calculado novamente o valor do m² por Bairro

dados_bairro["Valor_m2_bairro"] = dados_bairro["Valor_anuncio_float"]/dados_bairro["Metragem"]
dados_bairro
````

````python
dados_vendas.groupby("Bairro").mean()
````

````python
dados_bairro.loc["Vila Mariana"]
````

````python
dados_bairro.loc["Vila Mariana"]["Valor_m2_bairro"]
````

````python
dados_bairro["Valor_m2_bairro"]["Vila Mariana"]
````

````python
# Pegando os 10 bairros mais frequentes 

top_bairros = dados_vendas["Bairro"].value_counts()[:10].index
````

````python
# resetando o indice /// bairro precisa ser coluna e não indice

dados_bairro.reset_index(inplace=True)
dados_bairro
````

````python
# seleção dos top 10 bairros

dados_bairro.query("Bairro in @top_bairros")
````

````python
dados_bairro.head()
````

````python
# Gráfico usando os top 10 bairros

plt.figure(figsize=(14,6)) 

# plota o gráfico barplot com dados do @top_bairro
ax = sns.barplot(x="Bairro", y="Valor_m2_bairro", data=dados_bairro.query("Bairro in @top_bairros"))

# rotaciona o nome dos bairros no eixo x
ax.tick_params(axis='x', rotation=45) 
````

````python
#Gráfico tipo boxplot bairros x valor do anuncio (plotando valor)

plt.figure(figsize=(10,8))
ax = sns.boxplot(data = dados_vendas.query("Bairro in @top_bairros"), x="Bairro", y="Valor_anuncio_float")

# rotaciona o nome dos bairros no eixo x
ax.tick_params(axis='x', rotation=45) 

plt.show()
````

````python
# grafico boxplot com o 10 bairros excluindo metragem maior que 30 mil (plotando a metragem)

plt.figure(figsize=(10,8))

# plota gráfico com o top bairros menos metragem maior que 30 mil
ax = sns.boxplot(data = dados_vendas.query("Bairro in @top_bairros & Metragem < 30000"), x="Bairro", y="Metragem")

# rotaciona o nome dos bairros no eixo x
ax.tick_params(axis='x', rotation=45) 

plt.show()
````
