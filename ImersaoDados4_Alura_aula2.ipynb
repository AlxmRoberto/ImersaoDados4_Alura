#Desafio Aula 2

- Criar uma escala de valores em milhões
- Deixar o gráfico do histograma de valores legível, alterar as labels, cores, titulos, escala)
- Preço do metro quadrado por bairro e um gráfico
- Explorar as bibliotecas de visualizações e colocar as suas conclusões
- Pesquisar uma visualização para analisar as quartis, mediana e outliers

```python
# 1 - Criar uma escala de valores em milhoes (10^6)

dados_vendas['Valor_anuncio_milhoes'] = dados_vendas['Valor_anuncio_float'].apply(lambda i: i/(1000000))
dados_vendas.head()
```

```python
# 2 - Deixar o gráfico do histograma de valores legível (alterar labels, cores, título, escala)

plt.figure(figsize=(15, 8))
ax = srn.histplot(data = dados_vendas, x = "Valor_anuncio_milhoes", kde = True)
ax.set_title("Histograma Valor Imóvel (milhões)")
plt.xlim((-5, 20))
plt.xlabel('Valor (milhões)')
plt.ylabel('count')
plt.show()
```

```python
# 3 - Preço do metro quadrado por bairro e plotar em um gráfico ideal

dados_vendas["Preco_m2"] = (dados_vendas['Valor_anuncio_float'])/dados['Metragem']
dados_vendas.head()
```

```python
plt.figure(figsize=(10, 6))
ax = srn.histplot(data = dados_vendas, x = "Preco_m2", kde = True)
ax.set_title("Histograma Preço por m²")
plt.xlim((-5, 40000))
plt.xlabel('Preço (m²)')
plt.ylabel('count')
plt.show()
```
